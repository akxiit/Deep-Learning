```python
datagon = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2
)
```

```python
img = image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
i = 0

for output in datagon.flow(input_batch, batch_size=1,save_to_dir='/content/drive/MyDrive/Personal/Deep Learning/Data Augmentation/aug'):
  i = i + 1

  if i == 10:
    break
```
